```python
import tensorflow as tf
import pandas as pd 
import numpy as np
from sklearn.metrics import accuracy_score
import pickle
```

```python
train_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML_production/fashion-mnist-train-1.csv")
train_df2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML_production/fashion-mnist-train-2.csv")

test_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML_production/fashion-mnist_test.csv")
```

```python
x_train = train_df.drop("label", axis=1).values / 255.0
x_train2 = train_df2.drop("label", axis=1).values / 255.0

y_train = train_df["label"].values
y_train2 = train_df2["label"].values
```

```python
x_test = test_df.drop("label", axis=1).values / 255.0
y_test = test_df["label"].values
```

```python
# Build the model 
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28 * 28,)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Compile the model 
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
#model.fit(x_train, y_train, epochs=30)
model.fit(x_train2, y_train2, epochs=30)
```

```python
# Evaluate the model 1
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test Accuracy:', test_acc)
```

```python
model.save('/content/drive/MyDrive/Colab Notebooks/model.h5')
```

```python
# Evaluate the model 2
test_loss2, test_acc2 = model.evaluate(x_test, y_test)
print('Test Accuracy:', test_acc2)
```
